```python
!bedtools sort -faidx ./YOA-data/hg38.chrom.sizes \
              -i ./YOA-work2/CLIP_TIA1.bed \
              > ./YOA-work2/CLIP_TIA1_sorted.bed
```

```python
!bedtools sort -faidx ./YOA-data/hg38.chrom.sizes \
              -i ./YOA-work2/CLIP_TIA1_2.bed \
              > ./YOA-work2/CLIP_TIA1_2_sorted.bed
```

```python
%%bash
TMPDIR="/tmp"

INBED="./YOA-work2/CLIP_TIA1_2.bed"
OUTBED="./YOA-work2/CLIP_TIA1_2_sorted.bed"
CHROMS="./YOA-data/hg38.chrom.sizes"

> "$OUTBED"

while read chr len; do
  awk -v c="$chr" '$1==c' "$INBED" \
    | LC_ALL=C sort -k2,2n --parallel=4 -S 1G \
        --temporary-directory="$TMPDIR" \
    >> "$OUTBED"
done < "$CHROMS"
```

```python
!bedtools intersect -s -u -sorted -g ./YOA-data/hg38.chrom.sizes -a ./YOA-work2/CLIP_TIA1_2_sorted.bed -b ./YOA-work2/CLIP_TIA1_2_sorted.bed > ./YOA-work2/reproducible_TIA1_peaks.bed
```

```python
!bedtools intersect -s -a ./YOA-work2/reproducible_TIA1_peaks.bed -b gencode_UTR.bed -wb > ./YOA-work2/TIA1_gene_overlap.bed
```

```python
!awk '{split($10, a, "."); print a[1]}' ./YOA-work2/TIA1_gene_overlap.bed \
  | sort -u \
  > ./YOA-work2/TIA1_binding_genes.txt
```

```python
import mygene

with open('./YOA-work2/TIA1_binding_genes.txt', 'r') as f:
    enst_list = [line.strip().split('.')[0] for line in f if line.strip()] 

mg = mygene.MyGeneInfo()
query = mg.querymany(enst_list, scopes='ensembltranscript', fields='symbol', species='human')

symbol_list = [q['symbol'] for q in query if 'symbol' in q]

with open('./YOA-work2/TIA1_binding_genes_symbol.txt', 'w') as f:
    for sym in symbol_list:
        f.write(sym + '\n')

print(f'총 {len(symbol_list)}개의 gene symbol이 변환되어 저장되었습니다.')
```

```python
!sort -u ./YOA-work2/TIA1_binding_genes_symbol.txt > ./YOA-work2/TIA1_binding_genes_symbol_uniq.txt
```

```python
import gseapy as gp
print(gp.get_library_name(organism='Human'))
```

```python
enrich_go = gp.enrichr(
    gene_list='./YOA-work2/TIA1_binding_genes_symbol.txt',                
    gene_sets='GO_Biological_Process_2025',       
    organism='Human',                            
    outdir='gseapy_GO_results'                   
)
# KEGG 분석
enrich_kegg = gp.enrichr(
    gene_list='./YOA-work2/TIA1_binding_genes_symbol.txt',
    gene_sets='KEGG_2021_Human',
    organism='Human',
    outdir='gseapy_KEGG_results'
)
```

```python
from gseapy.plot import barplot

barplot(enrich_go.results.head(40), title='GO_BP Enrichment')
barplot(enrich_kegg.results.head(40), title='KEGG Pathway Enrichment')
```
